```python
import sqlite3

import pandas as pd
import numpy as np
```

```python
conn = sqlite3.connect('bls_data.db')
c = conn.cursor()
```

```python
c.execute("""SELECT * 
             FROM yoy""")
c.fetchone()
```

```python
msas = []
c.execute("""SELECT * 
             FROM yoy""")
data = {}
counter = 0
for i in c:
    print(i)
    y_id = i[0]
    msa = i[2]
    period = i[3]
    value = i[4]
    prev_value = i[5]
    change = i[6]
    data[counter] = pd.Series([y_id, msa, period, value, prev_value, change], 
                           index=['y_id', 'msa', 'period', 'value', 'prev_value', 'change'])
    counter+=1
df = pd.DataFrame(data).T
```

```python
def export_to_excel(data, file_name):
    writer = pd.ExcelWriter("%s.xlsx" % (file_name,))
    data.to_excel(writer,"%s" % (file_name,))
    writer.save()
    print('Done')
```

```python
export_to_excel(df, 'YOY')
```
